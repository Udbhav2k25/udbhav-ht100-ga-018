```python
# I am feeling very sad today.
# But I think things will get better soon.
# Actually I'm feeling happy now!
# Wait, I'm stressed again.
# Okay, I feel calm now.
from emotion_model import EmotionModel
from utils import create_emotion_graph
from IPython.display import Image

chat = [
    "I am feeling very sad today.",
    "But I think things will get better soon.",
    "Actually I'm feeling happy now!",
    "Wait, I'm stressed again.",
    "Okay, I feel calm now."
]

model = EmotionModel()

emotion_scores = []

for msg in chat:
    label, probs = model.predict_emotion(msg)
    emotion_scores.append(max(probs.values()))
    print(msg, "", label)

create_emotion_graph(emotion_scores)
Image("emotion_plot.png")
```
